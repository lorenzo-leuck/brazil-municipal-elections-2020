```python
idadeTw = pd.DataFrame(baseTw['NR_IDADE_DATA_POSSE'])
generoTw = pd.DataFrame(baseTw['DS_GENERO'].value_counts(normalize=True)*100)
instrucaoTw = pd.DataFrame(baseTw['DS_GRAU_INSTRUCAO'].value_counts(normalize=True)*100)
estadoCivilTw = pd.DataFrame(baseTw['DS_ESTADO_CIVIL'].value_counts(normalize=True)*100)
racaTw = pd.DataFrame(baseTw['DS_COR_RACA'].value_counts(normalize=True)*100)
ocupacaoTw = pd.DataFrame(baseTw['DS_OCUPACAO'].value_counts(normalize=True)*100)
partidosTw = pd.DataFrame(baseTw['SG_PARTIDO'].value_counts())
partidosTw.columns = ["Candidatos com Twitter"]
municipiosTw = pd.DataFrame(baseTw['MUNICIPIO_ESTADO'].value_counts())
municipiosTw.columns = ["Candidatos com Twitter"]
```

```python
idadeTw.median()
generoTw.head(1)
instrucaoTw.head(1)
estadoCivilTw.head(1)
racaTw.head(1)
ocupacaoTw.head(1)
temTwitterTw.head(1)
```

```python

distribuicaoCandidatosTw = pd.merge(partidos, partidosTw, left_index=True, right_index=True)
distribuicaoCandidatosTw['Candidatos sem Twitter'] = distribuicaoCandidatosTw['Candidatos'] - distribuicaoCandidatosTw['Candidatos com Twitter']

percentualCandidatosTw = distribuicaoCandidatosTw.copy()
percentualCandidatosTw['Candidatos sem Twitter'] = percentualCandidatosTw['Candidatos sem Twitter']/percentualCandidatosTw['Candidatos']

percentualCandidatosTw['Candidatos com Twitter'] = percentualCandidatosTw['Candidatos com Twitter']/percentualCandidatosTw['Candidatos']

percentualCandidatosTw = percentualCandidatosTw.drop(['Candidatos'], axis=1)

percentualCandidatosTw = percentualCandidatosTw.sort_values(by='Candidatos com Twitter', ascending=False)

percentualCandidatosTw

percentualCandidatosTw.to_clipboard(sep='\t')
```

```python
distribuicaoCandidatosTw = pd.merge(municipios, municipiosTw, left_index=True, right_index=True)
distribuicaoCandidatosTw['Candidatos sem Twitter'] = distribuicaoCandidatosTw['Candidatos'] - distribuicaoCandidatosTw['Candidatos com Twitter']

percentualCandidatosTw = distribuicaoCandidatosTw.copy()
percentualCandidatosTw['Candidatos sem Twitter'] = percentualCandidatosTw['Candidatos sem Twitter']/percentualCandidatosTw['Candidatos']

percentualCandidatosTw['Candidatos com Twitter'] = percentualCandidatosTw['Candidatos com Twitter']/percentualCandidatosTw['Candidatos']

percentualCandidatosTw = percentualCandidatosTw.drop(['Candidatos'], axis=1)

percentualCandidatosTw = percentualCandidatosTw.sort_values(by='Candidatos com Twitter', ascending=False)

percentualCandidatosTw

percentualCandidatosTw.to_clipboard(sep='\t')
```
